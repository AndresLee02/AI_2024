**Text Analysis Tool**: Create a Python program that reads a text file and performs various analyses on its content. The program should: Count the number of words, sentences, and paragraphs. Calculate the average word length. Identify the most common words and their frequencies.

**Hangman Game**: Create a Python program that allows a user to play the game of Hangman. The program should randomly select a word from a predefined list, and the player must guess letters one at a time until they guess the word correctly or run out of attempts.

**Number Guessing Game**: Create a program that generates a random number between 1 and 100, and then asks the user to guess the number. The program should provide hints to the user if the guessed number is greater or less than the target number, and keep asking for a new guess until the user guesses correctly.

```python
# TEXT ANALYSIS TOOL

def analyze_text_file(filename):
  """
  Analyzes a text file and prints various statistics about its content.

  Args:
    filename: The path to the text file.
  """

  # Here the model will read the content of the file
  with open(filename, "r") as file:
    text = file.read()

  # Here we split the text into words
  words = text.split()

  # Count the number of words
  num_words = len(words)

  # Now we split the text into sentences
  sentences = text.split(". ")

  # Count the number of sentences
  num_sentences = len(sentences)

  # And finally split the text into paragraphs
  paragraphs = text.split("\n\n")

  # Count the number of paragraphs
  num_paragraphs = len(paragraphs)

  # We calculate the average word length
  total_length = sum(len(word) for word in words)
  avg_word_length = total_length / num_words

  # Create a dictionary to store word frequencies
  word_counts = {}
  for word in words:
    word = word.lower()  # Convert words to lowercase for case-insensitive counting
    if word in word_counts:
      word_counts[word] += 1
    else:
      word_counts[word] = 1

  # Sort the word counts dictionary by frequency in descending order
  sorted_word_counts = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)

  print("\n Analysys results \n")

  # Print results
  print(f"File: {filename}")
  print("Number of words:", num_words)
  print("Number of sentences:", num_sentences)
  print("Number of paragraphs:", num_paragraphs)
  print("Average word length:", avg_word_length)
  print("Most common words:")
  for word, count in sorted_word_counts[:3]:  # Print the most common words
    print(f"- {word}: {count}")

# Example
filename = "exampleFile.txt"
analyze_text_file(filename)
```

```python
# HANGMAN GAME
import random

def get_word():

  words = ["python", "programming", "turing", "fime", "machine", "learning"]
  return random.choice(words)

def display_hangman(wrong_guesses):

  stages = [
      """
          --------
          |      |
          |
          |
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |      |
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |     /|
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |     /|/
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |     /|/
          |     /
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |     /|/
          |     / /
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      X
          |     /|/
          |     / /
          |
          |
          -------
      """
  ]
  return(stages[wrong_guesses])

def play_hangman():
  """
  Main function to play the Hangman game.
  """
  word = get_word().upper()
  word_letters = set(word)  # Convert word to letters
  alphabet = set('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
  wrong_guesses = 0
  guessed_letters = set()

  # Create an space for each letter
  word_display = ['_'] * len(word)

  print("Welcome to Hangman!")
  current_stage = display_hangman(wrong_guesses)
  print(current_stage)

  while wrong_guesses < len(current_stage) - 1 and len(word_letters) > 0:
    print("You have", len(alphabet), "letters left to guess.")


    print(" ".join(word_display))

    guess = input("Guess a letter: ").upper()

    if guess in alphabet:
      alphabet.remove(guess)
      if guess in word_letters:
        word_letters.remove(guess)

        for i in range(len(word)):
          if word[i] == guess:
            word_display[i] = guess
      else:
        wrong_guesses += 1
        print("Wrong guess! Try again.")
    else:
      print("Invalid guess. Please enter a single letter.")

    print(display_hangman(wrong_guesses))

  if wrong_guesses < len(current_stage) - 1:
    print(f"Congratulations! You guessed the word {word} correctly.")
  else:
    print(f"You ran out of guesses. The word was {word}.")

play_hangman()
```

```python
# NUMBER GUESSING GAME

import random

def play_guessing_game():

  correct_number = random.randint(1, 100)

  guess = None
  num_guesses = 0

  while guess != correct_number:

    try:
      guess = int(input("Guess a number between 1 and 100: "))
      num_guesses += 1
    except ValueError:
      print("Invalid input. Please enter a number.")
      continue

    # Give hints based on guess
    if guess < correct_number:
      print("Too low, try again!")
    elif guess > correct_number:
      print("Too high, try again!")

  print(f"Congratulations! You guessed the number in {num_guesses} tries.")

play_guessing_game()
```
